A NoSQL database is a non-relational database that is likely to be more document and distributed than structured. Although relational databases are reliable, there are different sorts of data that isn't a natural fit for a Relational format.

A document store is a database that stores records as unique documents in the database. In these document stores, each document contains key-value pairs, which makes the document stores flexible as each document can be unique.

Key-value stores work like a Python dictionary, and are powered by a hash table under the hood. To access data, we can simply pass the uique key for the desired value.

Column Stores work like pandas, with each column getting its own series. Each column is stored separately, allowing for very fast reads when querying data.

Graph databases excel sorting data where there are multiple associations between the entities sotred in the database, using nodes and edges. A good example is social media.

## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


Resilient Distributed Dataset (RDD) is a collection of an arbitrary number of server clusters that share a full dataset distributed between them. If the RDD is not fault torlerant, we could get incorrect information back because we think we've queried the entire dataset, but don't realize some servers were down. The RDDs such as Hadoop and Spark account for fault tolerance by chopping the data into different labeled blocks, and then storing multiple different backups for each block across all different server clusters.





## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
MapReduce is the idea of having each server run some query or function at the same time, and then combining the results. When we ask each server to run the function or query at the same time, this is the "Map" step. When we combine the results from each server into a single aggregate, this is the "Reduce" step.
